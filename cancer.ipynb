```python
a=pd.get_dummies(data["diagnosis"])
cancer=pd.concat([data,a],axis="columns")
cancer.drop(["diagnosis","B"],axis="columns",inplace=True)
cancer.rename(columns={"M":"Malignant/Benign"},inplace=True)
```

```python
data.head()
```

```python
Y=cancer[["Malignant/Benign"]]
X=cancer.drop(["Malignant/Benign"],axis="columns")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=42,test_size=0.2)
```

```python
print("x train length : ",len(x_train))
print("x test length : ",len(x_test))
print("y train length : ",len(y_train))
print("y test length : ",len(y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
prediction[:5]
```

```python
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score
print("mean_squared_error : ",mean_squared_error(y_test,prediction))
print("r2_score : ",r2_score(y_test,prediction))
```

```python
from sklearn.tree import DecisionTreeRegressor
model1 = DecisionTreeRegressor()
model1.fit(x_train,y_train)
```

```python
prediction1 = model1.predict(x_test)
prediction1[:5]
```

```python
print("mean_squared_error : ",mean_squared_error(y_test,prediction1))
print("r2_score : ",r2_score(y_test,prediction1))
```

```python
from sklearn.neighbors import KNeighborsRegressor
model2 = KNeighborsRegressor()
model2.fit(x_train,y_train)
```

```python
prediction2 = model2.predict(x_test)
prediction2[:5]
```

```python
print("mean_squared_error : ",mean_squared_error(y_test,prediction2))
print("r2_score : ",r2_score(y_test,prediction2))
```

```python
from sklearn.linear_model import LinearRegression
model3 = LinearRegression()
model3.fit(x_train,y_train)
```

```python
prediction3 = model3.predict(x_test)
prediction3[:5]
```

```python
print("mean_squared_error : ",mean_squared_error(y_test,prediction3))
print("r2_score : ",r2_score(y_test,prediction3))
```
