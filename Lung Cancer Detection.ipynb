# Step 5 - Splitting data into feature array amd target column¶

```python
X=data.drop(columns=["LUNG_CANCER","GENDER"]) #feature array
Y=data["LUNG_CANCER"]
```

```python
X.head()
```

```python
Y.head()
```

# Scaling of data (Optional)

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X=sc.fit_transform(X)
```

```python
X[:5]
```

# Step 6 - Splitting data into Testing & Training

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=42,test_size=0.2)
```

```python
print("x train length : ",len(x_train))
print("x test length : ",len(x_test))
print("y train length : ",len(y_train))
print("y test length : ",len(x_test))
```

# Step 7 - Building a model

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier()
model1.fit(x_train,y_train)
```

# Step 8 - Prediction using model

```python
prediction = model.predict(x_test)
prediction[:5]
```

```python
prediction1 = model1.predict(x_test)
prediction1[:5]
```

# Step 9 - Accuracy check of the model

```python
from sklearn.metrics import f1_score,precision_score,recall_score
print("f1_score = ",f1_score(y_true=y_test,y_pred=prediction,average="micro"))
print("precision_score = ",precision_score(y_true=y_test,y_pred=prediction,average="micro"))
print("recall_score = ",recall_score(y_true=y_test,y_pred=prediction,average="micro"))
```

```python
from sklearn.metrics import f1_score,precision_score,recall_score
print("f1_score = ",f1_score(y_true=y_test,y_pred=prediction1,average="micro"))
print("precision_score = ",precision_score(y_true=y_test,y_pred=prediction1,average="micro"))
```

# Step 10 - Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusionmat=confusion_matrix(y_test,prediction)
print(confusionmat)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
visualrep=ConfusionMatrixDisplay(confusion_matrix=confusionmat,display_labels=model.classes_)
visualrep.plot()
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```
